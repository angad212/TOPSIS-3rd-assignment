```python
!pip install transformers datasets rouge-score nltk seaborn psutil

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import time
import psutil
from datasets import load_dataset
from transformers import pipeline, AutoModelForCausalLM, AutoTokenizer
from rouge_score import rouge_scorer
from nltk.translate.bleu_score import sentence_bleu
import torch

dataset = load_dataset("wikitext", "wikitext-2-raw-v1", split="test")

models = {
    "GPT-2 Small": "distilgpt2",  # Use smaller model for faster inference
    "GPT-2": "gpt2",
    "GPT-2 Medium": "gpt2-medium"
}

results = []
scorer = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'], use_stemmer=True)
device = 0 if torch.cuda.is_available() else -1

for model_name, model_id in models.items():
    try:
        generator = pipeline("text-generation", model=model_id, device=device)
        tokenizer = AutoTokenizer.from_pretrained(model_id)
        model_size = sum(p.numel() for p in AutoModelForCausalLM.from_pretrained(model_id).parameters()) / 1e9
    except Exception as e:
        print(f"Error loading model {model_name}: {e}")
        continue

    for i in range(2):  # Limit to 2 samples per model for faster results
        prompt = dataset[i]['text'][:100]  # Shorten prompt for quicker results
        process = psutil.Process()

        try:
            start_time = time.time()
            start_mem = process.memory_info().rss / (1024 * 1024)
            output = generator(prompt, max_length=150, num_return_sequences=1)[0]['generated_text']
            end_time = time.time()
            end_mem = process.memory_info().rss / (1024 * 1024)
        except Exception as e:
            print(f"Error generating text with {model_name}: {e}")
            continue

        inference_time = end_time - start_time
        memory_usage = end_mem - start_mem
        reference = dataset[i]['text']
        bleu = sentence_bleu([reference.split()], output.split())
        rouge_scores = scorer.score(reference, output)

        results.append({
            "Model": model_name,
            "BLEU": bleu,
            "ROUGE-1": rouge_scores['rouge1'].fmeasure,
            "ROUGE-2": rouge_scores['rouge2'].fmeasure,
            "ROUGE-L": rouge_scores['rougeL'].fmeasure,
            "Inference Time": inference_time,
            "Model Size": model_size,
            "Memory Usage": memory_usage
        })

df = pd.DataFrame(results).groupby("Model").mean().reset_index()
weights = np.array([0.25, 0.20, 0.15, 0.10, 0.10, 0.10, 0.10])
beneficial = ['BLEU', 'ROUGE-1', 'ROUGE-2', 'ROUGE-L']
non_beneficial = ['Inference Time', 'Memory Usage']

normalized_df = df.copy()
for col in beneficial:
    normalized_df[col] = df[col] / np.sqrt((df[col] ** 2).sum())
for col in non_beneficial:
    normalized_df[col] = (df[col].max() - df[col]) / (df[col].max() - df[col].min() + 1e-9)

criteria = df.columns[1:]
weighted_df = normalized_df.copy()
for i, col in enumerate(criteria):
    weighted_df[col] *= weights[i]

ideal_best = weighted_df[criteria].max()
ideal_worst = weighted_df[criteria].min()
distance_best = np.sqrt(((weighted_df[criteria] - ideal_best) ** 2).sum(axis=1))
distance_worst = np.sqrt(((weighted_df[criteria] - ideal_worst) ** 2).sum(axis=1))

df['TOPSIS Score'] = distance_worst / (distance_best + distance_worst)
df = df.sort_values(by='TOPSIS Score', ascending=False)

plt.figure(figsize=(10, 5))
sns.barplot(x=df['Model'], y=df['TOPSIS Score'], palette='viridis')
for index, value in enumerate(df['TOPSIS Score']):
    plt.text(index, value + 0.02, f"{value:.2f}", ha='center')

plt.title('TOPSIS Score Ranking for Text Generation Models')
plt.xlabel('Model')
plt.ylabel('TOPSIS Score')
plt.ylim(0, 1)
plt.xticks(rotation=20)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()

metrics = ['BLEU', 'ROUGE-1', 'ROUGE-2', 'ROUGE-L']
df.set_index('Model')[metrics].plot(kind='bar', figsize=(12, 6), colormap='coolwarm', alpha=0.85)
plt.title('BLEU and ROUGE Score Comparison')
plt.ylabel('Score')
plt.xticks(rotation=20)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()

efficiency_metrics = ['Inference Time', 'Memory Usage']
df.set_index('Model')[efficiency_metrics].plot(kind='bar', figsize=(12, 6), colormap='magma', alpha=0.85)
plt.title('Efficiency Metrics Comparison')
plt.ylabel('Value')
plt.xticks(rotation=20)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()

df.to_csv('text_generation_topsis_results.csv', index=False)

from google.colab import files
files.download('text_generation_topsis_results.csv')
```

```python
!pip install transformers datasets rouge-score nltk seaborn psutil

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import time
import psutil
from datasets import load_dataset
from transformers import pipeline, AutoModelForCausalLM, AutoTokenizer
from rouge_score import rouge_scorer
from nltk.translate.bleu_score import sentence_bleu
import torch

dataset = load_dataset("wikitext", "wikitext-2-raw-v1", split="test")

models = {
    "GPT-2": "gpt2",
    "GPT-2 Medium": "gpt2-medium",
    "GPT-2 Large": "gpt2-large",
    "GPT-Neo": "EleutherAI/gpt-neo-1.3B",
    "OPT": "facebook/opt-1.3b",
    "BLOOM": "bigscience/bloom-560m"
}

results = []
scorer = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'], use_stemmer=True)
device = 0 if torch.cuda.is_available() else -1

for model_name, model_id in models.items():
    try:
        generator = pipeline("text-generation", model=model_id, device=device)
        tokenizer = AutoTokenizer.from_pretrained(model_id)
        model_size = sum(p.numel() for p in AutoModelForCausalLM.from_pretrained(model_id).parameters()) / 1e9
    except Exception as e:
        print(f"Error loading model {model_name}: {e}")
        continue

    for i in range(5):
        prompt = dataset[i]['text'][:100]
        process = psutil.Process()

        try:
            start_time = time.time()
            start_mem = process.memory_info().rss / (1024 * 1024)
            output = generator(prompt, max_length=150, num_return_sequences=1)[0]['generated_text']
            end_time = time.time()
            end_mem = process.memory_info().rss / (1024 * 1024)
        except Exception as e:
            print(f"Error generating text with {model_name}: {e}")
            continue

        inference_time = end_time - start_time
        memory_usage = end_mem - start_mem
        reference = dataset[i]['text']
        bleu = sentence_bleu([reference.split()], output.split())
        rouge_scores = scorer.score(reference, output)

        try:
            input_ids = tokenizer(output, return_tensors='pt').input_ids
            with torch.no_grad():
                model = AutoModelForCausalLM.from_pretrained(model_id)
                outputs = model(input_ids, labels=input_ids)
            loss = outputs.loss
            perplexity = torch.exp(loss).item()
        except Exception as e:
            print(f"Error calculating perplexity for {model_name}: {e}")
            perplexity = np.nan

        results.append({
            "Model": model_name,
            "BLEU": bleu,
            "ROUGE-1": rouge_scores['rouge1'].fmeasure,
            "ROUGE-2": rouge_scores['rouge2'].fmeasure,
            "ROUGE-L": rouge_scores['rougeL'].fmeasure,
            "Perplexity": perplexity,
            "Inference Time": inference_time,
            "Model Size": model_size,
            "Memory Usage": memory_usage
        })

df = pd.DataFrame(results).groupby("Model").mean().reset_index()
weights = np.array([0.25, 0.20, 0.15, 0.10, 0.10, 0.10, 0.10])
beneficial = ['BLEU', 'ROUGE-1', 'ROUGE-2', 'ROUGE-L']
non_beneficial = ['Perplexity', 'Inference Time', 'Memory Usage']

normalized_df = df.copy()
for col in beneficial:
    normalized_df[col] = df[col] / np.sqrt((df[col] ** 2).sum())
for col in non_beneficial:
    normalized_df[col] = (df[col].max() - df[col]) / (df[col].max() - df[col].min() + 1e-9)

criteria = df.columns[1:]
weighted_df = normalized_df.copy()
for i, col in enumerate(criteria):
    weighted_df[col] *= weights[i]

ideal_best = weighted_df[criteria].max()
ideal_worst = weighted_df[criteria].min()
distance_best = np.sqrt(((weighted_df[criteria] - ideal_best) ** 2).sum(axis=1))
distance_worst = np.sqrt(((weighted_df[criteria] - ideal_worst) ** 2).sum(axis=1))

df['TOPSIS Score'] = distance_worst / (distance_best + distance_worst)
df = df.sort_values(by='TOPSIS Score', ascending=False)

plt.figure(figsize=(10, 5))
sns.barplot(x=df['Model'], y=df['TOPSIS Score'], palette='viridis')
for index, value in enumerate(df['TOPSIS Score']):
    plt.text(index, value + 0.02, f"{value:.2f}", ha='center')

plt.title('TOPSIS Score Ranking for Text Generation Models')
plt.xlabel('Model')
plt.ylabel('TOPSIS Score')
plt.ylim(0, 1)
plt.xticks(rotation=20)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()

metrics = ['BLEU', 'ROUGE-1', 'ROUGE-2', 'ROUGE-L']
df.set_index('Model')[metrics].plot(kind='bar', figsize=(12, 6), colormap='coolwarm', alpha=0.85)
plt.title('BLEU and ROUGE Score Comparison')
plt.ylabel('Score')
plt.xticks(rotation=20)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()

efficiency_metrics = ['Perplexity', 'Inference Time', 'Memory Usage']
df.set_index('Model')[efficiency_metrics].plot(kind='bar', figsize=(12, 6), colormap='magma', alpha=0.85)
plt.title('Efficiency Metrics Comparison')
plt.ylabel('Value')
plt.xticks(rotation=20)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()

df.to_csv('text_generation_topsis_results.csv', index=False)

from google.colab import files
files.download('text_generation_topsis_results.csv')
```
